```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
def process_data(file_name):
    file = open(file_name, "r")
    epsilon_raw = file.read()
    spilt_index = epsilon_raw.index('validation')
    eps_str = epsilon_raw[6:spilt_index-2].split(', ')
    accuracy_str = epsilon_raw[spilt_index+17:len(epsilon_raw)-2].split(', ')
    eps = []
    accuracy = []
    for item in eps_str:
        eps.append(float(item))
    for acc in accuracy_str:
        accuracy.append(float(acc))
    return eps,accuracy
```

```python
eps_1_slr,accuracy_1_slr = process_data("mnist_dpsgd_delta_0.01_lr_3e-06.txt")
eps_2_slr,accuracy_2_slr = process_data("mnist_dpsgd_delta_0.001_lr_3e-06.txt")
eps_3_slr,accuracy_3_slr = process_data("mnist_dpsgd_delta_0.0001_lr_3e-06.txt")
eps_4_slr,accuracy_4_slr = process_data("mnist_dpsgd_delta_1e-05_lr_3e-06.txt")
plt.plot(range(100), eps_1_slr, label='delta=1e-02')
plt.plot(range(100), eps_2_slr, label='delta=1e-03')
plt.plot(range(100), eps_3_slr, label='delta=1e-04')
plt.plot(range(100), eps_4_slr, label='delta=1e-05')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Epsilon')
plt.title('epoch - Îµ relations')

fig = matplotlib.pyplot.gcf()
fig.set_size_inches(8, 6)
```
